# Current Population Survey K-Means Clustering

This dataset does not have many distinct clusters due to the nature of the survey data, but we will describe any clusters that correspond to categories we are trying to predict.

```python
import sklearn.cluster
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# Import dataset
data = pd.read_csv("../datasets/CPS2016_UPDATE.csv",
                  usecols=['age', 'weekly_hrs', 'educ', 
                'fam_income', 'num_in_house', 'num_child'])

# We are going to use the entire dataset for this model
# to test on a smaller sample uncomment below
# data = data.sample(n=1000)
```

```python
# Two Clusters
fig, ax = plt.subplots(6,6, figsize=(12,12), dpi=100)
for i, var1 in enumerate(data):
    for j, var2 in enumerate(data):
        X = data[[var1, var2]].values
        
        kmc = sklearn.cluster.KMeans(2, n_init=100)
        kmc.fit(X)
        cluster_labels = kmc.predict(X)
        cluster0 = (cluster_labels == 0)
        cluster1 = (cluster_labels == 1)

        ax[j,i].scatter(X[cluster0,0], X[cluster0,1], marker='x', color='red')
        ax[j,i].scatter(X[cluster1,0], X[cluster1,1], marker='+', color='blue')
        ax[j,i].set_xlabel(var1)
        ax[j,i].set_ylabel(var2)
fig.tight_layout()
plt.show(fig)
```

- Age and Education clusters correlate to groups of individuals who are married vs not married
- Education and Family Income clusters correlate to individuals who are married vs not married

```python
# Three Clusters
fig, ax = plt.subplots(6,6, figsize=(12,12), dpi=100)
for i, var1 in enumerate(data):
    for j, var2 in enumerate(data):
        X = data[[var1, var2]].values
        
        kmc = sklearn.cluster.KMeans(3, n_init=100)
        kmc.fit(X)
        cluster_labels = kmc.predict(X)  # will be 0 and 1 and 2
        cluster0 = (cluster_labels == 0)
        cluster1 = (cluster_labels == 1)
        cluster2 = (cluster_labels == 2)

        ax[j,i].scatter(X[cluster0,0], X[cluster0,1], marker='x', color='red')
        ax[j,i].scatter(X[cluster1,0], X[cluster1,1], marker='+', color='blue')
        ax[j,i].scatter(X[cluster2,0], X[cluster2,1], marker='o', color='orange')
        ax[j,i].set_xlabel(var1)
        ax[j,i].set_ylabel(var2)
fig.tight_layout()
plt.show(fig)
```
