# Plotting
It's good to plot as quickly as possible, so you can bring your human experience to the problem.
The main plotting library is matplotlib, which is powerful but contankerous.

```python
fig, ax = plt.subplots(1, figsize=(6,6), dpi=100)
ax.set_xlabel("weight") 
ax.set_ylabel("acceleration")
ax.scatter(data['weight'], data['acceleration'])
fig.tight_layout() # this makes it look a little better
plt.show(fig)
```

I want to plot across all of the variables, which is a good chance to see how to use some looping tools

```python
for var in data:
    print(var)
```

```python
for i, var in enumerate(data):
    print(i, var)
```

```python
fig, ax = plt.subplots(1,8, figsize=(12,3), dpi=100)
for i,var in enumerate(data):
        ax[i].scatter(data[var], data['mpg'])
        ax[i].set_xlabel(var)
        ax[i].set_ylabel(var)
fig.tight_layout()
plt.show(fig)
```

Maybe you want to plot each variable against every other variable, to see any correlations.
You could to this with two loops

```python
fig, ax = plt.subplots(8,8, figsize=(12,12), dpi=100)
for i, var1 in enumerate(data):
    for j, var2 in enumerate(data):
        ax[j,i].scatter(data[var1], data[var2])
        ax[j,i].set_xlabel(var1)
        ax[j,i].set_ylabel(var2)
fig.tight_layout()
plt.show(fig)
```

But, this is something people like to do a lot, so Pandas has a builtin command for it!

```python
pd.plotting.scatter_matrix(data, figsize=(12,12))
plt.show()
```

# Building a model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(data[['displacement', 'cylinders', 'weight']], data['mpg'])
```

```python
print(model.coef_)
print(model.intercept_)
print(f"The prediction model is:\nsales =   {model.coef_[0]}*(Displacement)\n\t+ {model.coef_[1]}*(cylinders)\n\t+ {model.coef_[2]}*(weight)\n\t+ {model.intercept_}")
```

## Please note!  What I am about to do (evaluating on the training data) is scientifically very very bad.  For now, I just want you to see how the function work.  We will do things "correctly" soon.

```python
predicted_sales = model.predict(data[['displacement', 'cylinders', 'weight']])
```

```python
fig, ax = plt.subplots(1,8, figsize=(12,3), dpi=100)
for i,var in enumerate(data):
        ax[i].scatter(data[var], data['mpg'], color='blue')
        ax[i].scatter(data[var], predicted_sales, color='red')
        ax[i].set_xlabel(var)
        ax[i].set_ylabel("mpg")
        ax[i].set_title("Actual vs Predicted")
fig.tight_layout()
plt.show(fig)
```

```python
residual = predicted_sales - data['mpg']
RSS = np.sum(residual**2)
print(RSS)

DSS = np.sum((data['mpg'] - data['mpg'].mean())**2)
print(DSS)

R2 = 1 - (RSS/DSS)
print(R2)
```

```python
model.score(data[['displacement', 'cylinders', 'weight']], data['mpg'])
```
